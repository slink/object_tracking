# Image Manipulation with skimage

This example builds a simple UI for performing basic image manipulation with [scikit-image](http://scikit-image.org/).

```python
from IPython.html.widgets import interact, interactive, fixed, ButtonWidget
from IPython.display import display
from skimage import data, filter, io
import numpy as np
from matplotlib import pylab as plt

from helper_functions import test_image_list, background_image_list
image_folder = 'test_jpgs'
%matplotlib inline
```

```python
image_list = background_image_list(image_folder, 1200)
im = plt.imread(image_list[0])
im1 = plt.imread(image_list[100])
plt.imshow(im)
plt.axis('off')
```

```python
def edit_image(image, sigma=0.1, r=1.0, g=1.0, b=1.0):
    new_image = filter.gaussian_filter(image, sigma=sigma, multichannel=True)
    new_image[:,:,0] = r*new_image[:,:,0]
    new_image[:,:,1] = g*new_image[:,:,1]
    new_image[:,:,2] = b*new_image[:,:,2]
    new_image = io.Image(new_image)
    display(new_image)
    return new_image

def add_lines(image, line1 = 100, line2 = 200):
  new_image = image.copy()  # Work on a copy
  new_image[line1,:,:] = 0
  new_image[line2,:,:] = 0
  new_image = io.Image(new_image)
  display(new_image)
```

```python
lims = (0,400,2)

w = interactive(add_lines, image=fixed(im), line1 = lims, line2 = lims)
display(w)

def on_limits_button_clicked(b):
    print [x.value for x in w.children]
    
def on_advace_button_clicked(b):
    w.kwargs['image'] = data.coffee()

limits_button = ButtonWidget(description="Top and Bottom Limits")
limits_button.on_click(on_limits_button_clicked)
display(limits_button)

advance_button = ButtonWidget(description="Next Image")
advance_button.on_click(on_advace_button_clicked)
display(advance_button)
```

## Python 3 only: Function annotations

```python
lims = (0.0,1.0,0.01)

@interact
def edit_image(image: fixed(i), sigma:(0.0,10.0,0.1)=0.1, r:lims=1.0, g:lims=1.0, b:lims=1.0):
    new_image = filter.gaussian_filter(image, sigma=sigma, multichannel=True)
    new_image[:,:,0] = r*new_image[:,:,0]
    new_image[:,:,1] = g*new_image[:,:,1]
    new_image[:,:,2] = b*new_image[:,:,2]
    new_image = io.Image(new_image)
    display(new_image)
    return new_image
```
