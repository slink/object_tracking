```python
def find_front_edge(image):
    """
    find_front_edge(image)
    
    takes in a thresholded and filtered image and returns an (x,y) 
    pair of values corresponding to the front edge of the moving 'blob'
    """
    front_col = 0
    for i, column in enumerate(image.T):
        if column.max() == True: 
            front_col = i
            break
            
    row_up = image[:,front_col].argmax()
    row_down = image[:,front_col][::-1].argmax()
    front_row = int(np.mean([row_up, row_down]))
    return (front_col, front_row)
```

```python
%pylab inline
from skimage import data, filter, color
from skimage.transform import hough_circle
from skimage.feature import peak_local_max
from skimage.draw import circle_perimeter
from skimage.util import img_as_ubyte
from scipy.misc import imread
```

```python
photo1 = imread('IMG_20140219_083446.jpg', flatten = True)
photo2 = imread('IMG_20140219_083456.jpg', flatten = True)
```

```python
# Load picture and detect edges
edges = filter.canny(photo1.T, sigma=20.0,
                     low_threshold=0.55, high_threshold=0.8)
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(8, 8))
ax[0].imshow(edges, cmap=plt.cm.gray)
ax[1].imshow(photo1.T, cmap=plt.cm.gray)
```

```python
# Detect two radii
hough_radii = np.arange(15, 30, 2)
hough_res = hough_circle(edges, hough_radii)

centers = []
accums = []
radii = []

for radius, h in zip(hough_radii, hough_res):
    # For each radius, extract two circles
    peaks = peak_local_max(h, num_peaks=2)
    centers.extend(peaks)
    accums.extend(h[peaks[:, 0], peaks[:, 1]])
    radii.extend([radius, radius])

# Draw the most prominent 5 circles
photo1 = color.gray2rgb(photo1)
for idx in np.argsort(accums)[::-1][:5]:
    center_x, center_y = centers[idx]
    radius = radii[idx]
    cx, cy = circle_perimeter(center_y, center_x, radius)
    photo1[cy, cx] = (220, 20, 20)

ax.imshow(photo1, cmap=plt.cm.gray)
```
